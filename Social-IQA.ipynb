```python
# Imports
import torch
import numpy as np
from torch.utils.data import DataLoader, Subset
from sklearn.metrics import accuracy_score, f1_score
from datasets import load_dataset
from transformers import (
    DistilBertTokenizerFast,
    DistilBertForMultipleChoice,
    DataCollatorWithPadding,
)
import matplotlib.pyplot as plt

# 1) Load dataset and tokenizer
dataset = load_dataset("allenai/social_i_qa")
tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")

# 2) Preprocessing function
def preprocess(example):
    inputs = []
    for choice in [example["answerA"], example["answerB"], example["answerC"]]:
        text = (
            f"Question: {example['question']} "
            f"Context: {example['context']} "
            f"Answer: {choice}"
        )
        tokenized = tokenizer(text, truncation=True, max_length=128)
        inputs.append(tokenized)

    return {
        "input_ids": [inp["input_ids"] for inp in inputs],
        "attention_mask": [inp["attention_mask"] for inp in inputs],
        "labels": int(example["label"]) - 1,
    }

train_data = dataset["train"].map(preprocess, remove_columns=dataset["train"].column_names)
val_data = dataset["validation"].map(preprocess, remove_columns=dataset["validation"].column_names)

# 4) Custom collator for multiple-choice
class MultipleChoiceCollator:
    def __init__(self, tokenizer):
        self.tokenizer = tokenizer

    def __call__(self, features):
        batch_size = len(features)
        num_choices = len(features[0]["input_ids"])

        flat_input_ids = []
        flat_attention_masks = []
        for f in features:
            flat_input_ids.extend(f["input_ids"])
            flat_attention_masks.extend(f["attention_mask"])

        padded = self.tokenizer.pad(
            {"input_ids": flat_input_ids, "attention_mask": flat_attention_masks},
            padding=True,
            return_tensors="pt",
        )

        input_ids = padded["input_ids"].view(batch_size, num_choices, -1)
        attention_mask = padded["attention_mask"].view(batch_size, num_choices, -1)
        labels = torch.tensor([f["labels"] for f in features], dtype=torch.long)

        return {
            "input_ids": input_ids,
            "attention_mask": attention_mask,
            "labels": labels
        }

data_collator = MultipleChoiceCollator(tokenizer)

def create_dataloader(data, indices, batch_size=16):
    subset = Subset(data, indices)
    return DataLoader(subset, batch_size=batch_size, collate_fn=data_collator)

def train_model(model, dataloader, optimizer, device):
    model.train()
    total_loss = 0.0
    for batch in dataloader:
        optimizer.zero_grad()
        outputs = model(
            input_ids=batch["input_ids"].to(device),
            attention_mask=batch["attention_mask"].to(device),
            labels=batch["labels"].to(device)
        )
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(dataloader)

def evaluate_model(model, dataloader, device):
    model.eval()
    all_preds, all_labels = [], []
    total_loss = 0.0
    with torch.no_grad():
        for batch in dataloader:
            outputs = model(
                input_ids=batch["input_ids"].to(device),
                attention_mask=batch["attention_mask"].to(device),
                labels=batch["labels"].to(device)
            )
            total_loss += outputs.loss.item()
            preds = torch.argmax(outputs.logits, dim=1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(batch["labels"].cpu().numpy())
    accuracy = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average="weighted")
    return total_loss / len(dataloader), accuracy, f1

initial_labeled = 5000
budget = 200
cycles = 10
patience = 3
indices = np.random.permutation(len(train_data))
labeled_idx = indices[:initial_labeled]
unlabeled_idx = indices[initial_labeled:]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
results = {"random": [], "margin": [], "least": []}

for strategy in ["margin", "least", "random"]:
    current_labeled_idx = labeled_idx.copy()
    current_unlabeled_idx = unlabeled_idx.copy()

    for cycle in range(cycles):
        train_loader = create_dataloader(train_data, current_labeled_idx)
        val_loader = create_dataloader(val_data, range(len(val_data)))

        model = DistilBertForMultipleChoice.from_pretrained("distilbert-base-uncased").to(device)
        optimizer = torch.optim.AdamW(model.parameters(), lr=3e-5)
        best_val_loss = float("inf")
        best_model_state = None
        no_improvement_count = 0

        for epoch in range(10):
            train_loss = train_model(model, train_loader, optimizer, device)
            val_loss, val_acc, val_f1 = evaluate_model(model, val_loader, device)
            print(f"[{strategy.upper()}] Cycle {cycle+1}, Epoch {epoch+1} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f} | Val F1: {val_f1:.4f}")
            if val_loss < best_val_loss:
                best_val_loss = val_loss
                best_model_state = model.state_dict().copy()
                torch.save(best_model_state, "best_model.pth")
                no_improvement_count = 0
            else:
                no_improvement_count += 1
            if no_improvement_count >= patience:
                print(f"Early stopping triggered after {epoch+1} epochs.")
                break

        model.load_state_dict(torch.load("best_model.pth"))
        val_loss, val_acc, val_f1 = evaluate_model(model, val_loader, device)
        results[strategy].append((val_acc, val_f1))

        if strategy in ["margin", "least"]:
            unlabeled_loader = create_dataloader(train_data, current_unlabeled_idx)
            scores = []
            with torch.no_grad():
                for batch in unlabeled_loader:
                    logits = model(
                        input_ids=batch["input_ids"].to(device),
                        attention_mask=batch["attention_mask"].to(device)
                    ).logits
                    probs = torch.softmax(logits, dim=1)
                    if strategy == "margin":
                        top2 = torch.topk(probs, 2).values.cpu().numpy()
                        scores.extend(top2[:, 0] - top2[:, 1])  # margin
                    else:
                        max_conf = torch.max(probs, dim=1).values.cpu().numpy()
                        scores.extend(max_conf)  # least confidence
            selected_indices = np.argsort(scores)[:budget] if strategy == "least" else np.argsort(scores)[:budget]
        else:
            selected_indices = np.random.choice(len(current_unlabeled_idx), budget, replace=False)

        new_labeled_idx = current_unlabeled_idx[selected_indices]
        current_labeled_idx = np.concatenate([current_labeled_idx, new_labeled_idx])
        current_unlabeled_idx = np.delete(current_unlabeled_idx, selected_indices)

# Plot results
x = np.arange(1, cycles + 1)
plt.figure(figsize=(10, 6))
for strategy in results:
    accs = [acc for acc, f1 in results[strategy]]
    plt.plot(x, accs, label=f"{strategy} - Acc")
plt.title("Validation Accuracy over Active Learning Cycles")
plt.xlabel("Cycle")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
for strategy in results:
    f1s = [f1 for acc, f1 in results[strategy]]
    plt.plot(x, f1s, label=f"{strategy} - F1")
plt.title("Validation F1 Score over Active Learning Cycles")
plt.xlabel("Cycle")
plt.ylabel("F1 Score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot results with adjusted y-axis limits
x = np.arange(1, cycles + 1)

# Find global min/max across all strategies for consistent scaling
all_accs = [acc for strategy in results for acc, f1 in results[strategy]]
all_f1s = [f1 for strategy in results for acc, f1 in results[strategy]]

# Calculate padding (5% of range)
acc_padding = (max(all_accs) - min(all_accs)) * 0.05
f1_padding = (max(all_f1s) - min(all_f1s)) * 0.05

# Accuracy plot
plt.figure(figsize=(10, 6))
for strategy in results:
    accs = [acc for acc, f1 in results[strategy]]
    plt.plot(x, accs, label=f"{strategy} - Acc", linewidth=2)
plt.title("Validation Accuracy over Active Learning Cycles")
plt.xlabel("Cycle")
plt.ylabel("Accuracy")
plt.ylim(min(all_accs) - acc_padding, max(all_accs) + acc_padding)  # Key change
plt.legend()
plt.grid(True)
plt.show()

# F1 Score plot
plt.figure(figsize=(10, 6))
for strategy in results:
    f1s = [f1 for acc, f1 in results[strategy]]
    plt.plot(x, f1s, label=f"{strategy} - F1", linewidth=2)
plt.title("Validation F1 Score over Active Learning Cycles")
plt.xlabel("Cycle")
plt.ylabel("F1 Score")
plt.ylim(min(all_f1s) - f1_padding, max(all_f1s) + f1_padding)  # Key change
plt.legend()
plt.grid(True)
plt.show()
```
